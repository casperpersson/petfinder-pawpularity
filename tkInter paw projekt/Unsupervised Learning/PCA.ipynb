```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_breast_cancer

bc = load_breast_cancer()
# Perform PCA on the breast cancer dataset
pca_bc = PCA(n_components=2)
pca_bc_result = pca_bc.fit_transform(bc.data)

# Scatter plot of the PCA results
plt.figure(figsize=(8, 8))
plt.scatter(pca_bc_result[:, 0], pca_bc_result[:, 1], c=bc.target, cmap='viridis', edgecolor='k')
plt.colorbar(label='Target')
plt.title('PCA of Breast Cancer Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
